## Setup

The cell below loads all required libraries and establishes the connection to the `petsdb` database.

---

```python
# Load Library
import sqlite3

# Connect to DB
conn = sqlite3.connect('petsdb')
cursor = conn.cursor()

# Verify connection to DB


def is_connected(conn):
    """Validates that the SQLite database is connected to petsdb."""
    try:
        conn.execute('SELECT * FROM persons LIMIT 1')
        return True
    except sqlite3.ProgrammingError as e:
        print(f'❌ Connection failed: {e}')
        return False


print(f'🔌 Database connection open: {is_connected(conn)}')
```

---
## 📋 Overview

This activity connects to the `petsdb` SQLite database and executes a series of SQL queries against two related tables:

- **persons** — contains `id`, `first_name`, `last_name`, `age`, `city`, and `zip_code`
- **pets** — contains `pet_name`, `pet_type`, `treatment_done`, and `owner_id` (foreign key → `persons.id`)

## 🔍 Exercises

1. What is the count of people belonging to different age groups in the persons table?
2. Which age group has the maximum number of people?
3. How many people do not have a last name?
4. How many people have more than one pet?
5. How many pets have received treatment?
6. How many pets have received treatment, and the type of pet is known?
7. How many pets are from the city called East Port?
8. How many pets are from the city called East Port, and who received treatment?

---

```python
# Q1: What is the count of people belonging to different age groups in the
# persons table?

cursor.execute(
    'SELECT age, COUNT(*) AS count FROM persons GROUP BY age ORDER BY age')
results = cursor.fetchall()

print('👥 Q1 — Age Group Counts:')
for row in results:
    print(f'   Age {row[0]:>3} → {row[1]:,} people')
```

```python
# Q2: Which age group has the maximum number of people?

cursor.execute(
    'SELECT age, COUNT(*) AS count FROM persons GROUP BY age ORDER BY count DESC LIMIT 1')
result = cursor.fetchone()

print(
    f'🏆 Q2 — The age group with the most people is Age {result[0]} with {result[1]:,} people.')
```

```python
# Q3: How many people do not have a last name?

cursor.execute('SELECT COUNT(*) FROM persons WHERE last_name IS NULL')
result = cursor.fetchone()

print(f'❓ Q3 — {result[0]:,} people do not have a last name on record.')
```

```python
# Q4: How many people have more than one pet?

cursor.execute(
    'SELECT COUNT(*) FROM (SELECT owner_id FROM pets GROUP BY owner_id HAVING COUNT(*) > 1)')
result = cursor.fetchone()

print(f'🐾 Q4 — {result[0]:,} people have more than one pet.')
```

```python
# Q5: How many pets have received treatment?

cursor.execute('SELECT COUNT(*) FROM pets WHERE treatment_done = 1')
result = cursor.fetchone()

print(f'💊 Q5 — {result[0]:,} pets have received treatment.')
```

```python
# Q6: How many pets have received treatment, and the type of pet is known?

cursor.execute(
    'SELECT COUNT(*) FROM pets WHERE treatment_done = 1 AND pet_type IS NOT NULL')
result = cursor.fetchone()

print(
    f'🐶🐱 Q6 — {result[0]:,} pets have received treatment and have a known pet type.')
```

```python
# Q7: How many pets are from the city called East Port?

cursor.execute(
    'SELECT COUNT(*) FROM pets JOIN persons ON pets.owner_id = persons.id WHERE LOWER(persons.city) = "east port"')
result = cursor.fetchone()

print(f'🏙️ Q7 — {result[0]:,} pets belong to owners from East Port.')
```

```python
# Q8: How many pets are from the city called East Port, and who received
# treatment?

cursor.execute('SELECT COUNT(*) FROM pets JOIN persons ON pets.owner_id = persons.id WHERE LOWER(persons.city) = "east port" AND pets.treatment_done = 1')
result = cursor.fetchone()

print(f'💊🏙️ Q8 — {result[0]:,} pets from East Port have received treatment.')
```
