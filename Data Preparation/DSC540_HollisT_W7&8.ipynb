# Chapter 7 — Data Cleaning and Transformation (Part 1)

In this first part of Chapter 7, I focus on the foundational cleaning steps using the MetObjects dataset. These operations help identify missing values, remove incomplete rows, fill in gaps with appropriate defaults, and eliminate duplicate records. Together, they establish a clean and reliable starting point for the transformations that follow.

### Steps for this section (1–4)

1. Use `.isna().sum()` to identify which columns contain missing values and understand the overall completeness of the dataset.  
2. Use `.dropna()` to filter out rows that are incomplete based on specific fields that are required for analysis.  
3. Use `.fillna()` to replace missing values with defaults such as `'Unknown'` for text fields or `0` for numeric fields.  
4. Use `.drop_duplicates()` to remove duplicate records based on a key column to ensure each entry is unique.

<hr style="border: none; height: 3px; background-color: black;">

```python
# 1: Identify missing values across all columns
missing_counts = met.isna().sum()
print('Missing values per column:')
print(missing_counts)

# 2: Filter out rows missing key descriptive fields
before_rows = len(met)

# Remove rows missing the artist name
met_no_missing_artist = met.dropna(subset=['Artist Display Name'])
after_artist_rows = len(met_no_missing_artist)

# Remove rows missing both Object Name and Title
met_no_missing_obj_title = met.dropna(
    subset=['Object Name', 'Title'], how='all')
after_obj_title_rows = len(met_no_missing_obj_title)

print(
    '\nRows removed due to missing Artist Display Name:',
    before_rows -
    after_artist_rows)
print('Rows removed due to missing Object Name and Title:',
      before_rows - after_obj_title_rows)

# 3: Fill in missing values with appropriate defaults
met_filled = met.copy()

before_culture_missing = met_filled['Culture'].isna().sum()
before_objectid_missing = met_filled['Object ID'].isna().sum()

# Replace missing Culture with 'Unknown' and missing Object ID with 0
met_filled['Culture'] = met_filled['Culture'].fillna('Unknown')
met_filled['Object ID'] = met_filled['Object ID'].fillna(0)

after_culture_missing = met_filled['Culture'].isna().sum()
after_objectid_missing = met_filled['Object ID'].isna().sum()

print(
    '\nFilled missing Culture values:',
    before_culture_missing -
    after_culture_missing)
print(
    'Filled missing Object ID values:',
    before_objectid_missing -
    after_objectid_missing)

# 4: Remove duplicate records based on Object ID
before_dupes = len(met)
met_no_duplicates = met.drop_duplicates(subset=['Object ID'])
after_dupes = len(met_no_duplicates)

print('\nDuplicate rows removed based on Object ID:', before_dupes - after_dupes)
```

# Chapter 7 — Data Cleaning and Transformation (Part 2)

This part of Chapter 7 focuses on transforming values, replacing specific entries, creating bins for numeric fields, and working with text fields. These operations help standardize the dataset, extract meaningful features, and prepare the data for grouping, visualization, and later analysis.

### Steps for this section

1. Use `.map()` and `.apply()` to transform values or apply custom functions to selected columns.  
2. Use `.replace()` to update specific values or apply a dictionary of replacements for consistency.  
3. Use `pd.cut()` and `pd.qcut()` to create equal‑width or quantile‑based bins for numeric fields.  
4. Use string methods such as `.str.upper()`, `.str.contains()`, and `.str.split()` to clean and extract information from text fields.

<hr style="border: none; height: 3px; background-color: black;">

```python
# 1. Transform values using mapping and a custom function

# Map the boolean public-domain flag to human-readable labels
pub_map = {True: 'Public', False: 'Restricted'}
met['PublicStatus'] = met['Is Public Domain'].map(pub_map)

# Extract numeric year values from the AccessionYear field


def extract_year(val):
    try:
        return int(val)
    except BaseException:
        return None


met['AccessionYear_Extracted'] = met['AccessionYear'].apply(extract_year)

mapped_count = met['PublicStatus'].notna().sum()
year_extracted_count = met['AccessionYear_Extracted'].notna().sum()

print('Values mapped for PublicStatus:', mapped_count)
print('AccessionYear values successfully extracted:', year_extracted_count)

# 2. Replace values for consistency across categories
met_replaced = met.copy()

# Replace 'Unknown' culture values with a clearer label
before_unknown = (met_replaced['Culture'] == 'Unknown').sum()
met_replaced['Culture'] = met_replaced['Culture'].replace(
    'Unknown', 'Not Specified')
after_unknown = (met_replaced['Culture'] == 'Not Specified').sum()

# Replace selected department names using a mapping dictionary
dept_map = {'European Paintings': 'Euro Paintings', 'Asian Art': 'Asia Dept'}
before_dept = met_replaced['Department'].isin(dept_map.keys()).sum()
met_replaced['Department'] = met_replaced['Department'].replace(dept_map)
after_dept = met_replaced['Department'].isin(dept_map.values()).sum()

print('\nCulture values replaced:', after_unknown)
print('Department values replaced:', after_dept)

# 3. Create bins for numeric fields using equal-width and quantile-based
# methods
met_bins = met.copy()

met_bins['ID_width_bins'] = pd.cut(met_bins['Object ID'], bins=4)
met_bins['ID_quantile_bins'] = pd.qcut(
    met_bins['Object ID'], q=4, duplicates='drop')

width_bin_count = met_bins['ID_width_bins'].notna().sum()
quantile_bin_count = met_bins['ID_quantile_bins'].notna().sum()

print('\nRows assigned to width-based bins:', width_bin_count)
print('Rows assigned to quantile-based bins:', quantile_bin_count)

# 4. Work with text fields to extract and standardize information
met_strings = met.copy()

# Convert artist names to uppercase for consistency
met_strings['Artist_Upper'] = met_strings['Artist Display Name'].str.upper()

# Extract the first word from the Medium field
met_strings['Medium_FirstWord'] = met_strings['Medium'].str.split().str[0]

# Identify rows where the Medium description contains the word "wood"
met_strings['Wood_Count'] = met_strings['Medium'].str.contains(
    'wood', case=False, na=False)

artist_upper_count = met_strings['Artist_Upper'].notna().sum()
medium_firstword_count = met_strings['Medium_FirstWord'].notna().sum()
wood_count = met_strings['Wood_Count'].sum()

print('\nArtist names converted to uppercase:', artist_upper_count)
print('Medium first words extracted:', medium_firstword_count)
print('Rows containing the word "wood":', wood_count)
```

This section focused on transforming values, replacing specific entries, creating bins for numeric fields, and working with text fields. These operations helped standardize the dataset, extract useful features, and prepare the data for later grouping and analysis.

**Value Transformations**

- AccessionYear values successfully converted to integers: **481,054** rows updated  

These transformations make the fields easier to interpret and more consistent for filtering and classification.

**Value Replacements**

Replacing values helps reduce category fragmentation and ensures consistent labeling across the dataset.

**Numeric Binning**

Both binning methods successfully categorized every row, confirming that Object ID contains no missing values and is suitable for numeric segmentation.

### Chapter 8 — Indexing and Merging (Part 1)

In this first part of Chapter 8, I focus on creating a hierarchical index and beginning the process of combining datasets from multiple sources. These steps help standardize the survey data, extract metadata from the PDF, and prepare the 2016 and 2017 datasets for reshaping and merging.

### Steps for this section:

1. Display all unique country variations and apply light standardization to improve consistency before indexing and merging.  
2. Create a hierarchical index using demographic fields from the 2016 survey dataset.  
3. Extract survey question metadata from the 2017 PDF file and convert it into a small DataFrame.  
4. Reshape the 2017 survey responses and merge them with the PDF metadata and the candy reference dataset to create a combined view of the data.

These steps demonstrate how to integrate multiple datasets, handle different formats, and build relationships between files.

<hr style="border: none; height: 3px; background-color: black;">

```python
# 1. Display all unique country variations before standardizing
unique_countries_before = df_2017['Q4: COUNTRY'].unique()
print('Unique country values before cleaning:')
print(unique_countries_before)
print(
    f'\nNumber of unique country variations before cleaning: {len(unique_countries_before)}')

# Light standardization pass: trim, uppercase, and normalize common US variants
df_2016['Which country do you live in?'] = (
    df_2016['Which country do you live in?']
    .astype(str)
    .str.strip()
    .str.upper()
    .replace({'US': 'USA', 'U.S.A': 'USA', 'U.S.A.': 'USA'})
)

df_2017['Q4: COUNTRY'] = (
    df_2017['Q4: COUNTRY']
    .astype(str)
    .str.strip()
    .str.upper()
    .replace({'US': 'USA', 'U.S.A': 'USA', 'U.S.A.': 'USA'})
)

# Show unique values after the initial cleanup
unique_countries_after = df_2017['Q4: COUNTRY'].unique()
print('\nUnique country values after initial cleaning:')
print(unique_countries_after)
print(
    f'\nNumber of unique country variations after initial cleaning: {len(unique_countries_after)}')

# ───────────────────────────────────────────────────────────────
# Full standardization: consolidate US, UK, and Canada variations
# ───────────────────────────────────────────────────────────────
print("\n=== Final Standardization of US, UK, and Canada Variations ===")

# Lists of messy variations to normalize
us_variations = [
    'USA',
    'u.s.',
    'UNITED STATES',
    'UNITED STAES',
    'UNITED STATES OF AMERICA',
    'USAUSAUSA',
    'AMERICA',
    'UNHINGED STATES',
    'US OF A',
    'UNITES STATES',
    'THE UNITED STATES',
    'NORTH CAROLINA',
    'UNIED STATES',
    'U S',
    'THE UNITED STATES OF AMERICA',
    'UNITE STATES',
    'INSANITY LATELY',
    'USA HARD TO TELL ANYMORE…',
    'MERICA',
    'USAS',
    'PITTSBURGH',
    'UNITED STATE',
    'NEW YORK',
    'TRUMPISTAN',
    'UNITED SATES',
    'CALIFORNIA',
    'I PRETEND TO BE FROM CANADA BUT I AM FROM THE UNITED STATES',
    'UNITED STATED',
    'AHEM...AMERCA',
    'NEW JERSEY',
    'UNITED STSTES',
    'UNITED STATSS',
    'MURRIKA',
    'USA! USA! USA!',
    'USAA',
    'ALASKA',
    'USSA',
    'U S A',
    'UNITED STATEA',
    'USA USA USA!!!!',
    'UD',
    'A']

uk_variations = ['UK', 'UNITED KINGDOM', 'ENGLAND', 'U.K.', 'ENDLAND']
canada_variations = ['CANADA', 'CAN', 'CANAE', 'CANADA`', 'CAN CANAE']

# Build mapping dictionaries (lowercased for matching)
us_map = {v.lower().strip(): 'U.S.A.' for v in us_variations}
uk_map = {v.lower().strip(): 'United Kingdom' for v in uk_variations}
canada_map = {v.lower().strip(): 'Canada' for v in canada_variations}

# Combine all mappings
country_map = {**us_map, **uk_map, **canada_map}

# Apply full standardization to both datasets
df_2017['Q4: COUNTRY'] = (
    df_2017['Q4: COUNTRY']
    .astype(str)
    .str.lower()
    .str.strip()
    .replace(country_map)
)

df_2016['Which country do you live in?'] = (
    df_2016['Which country do you live in?']
    .astype(str)
    .str.lower()
    .str.strip()
    .replace(country_map)
)

# Verification counts
print("\nTop 15 country values AFTER full standardization (2017):")
print(df_2017['Q4: COUNTRY'].value_counts().head(15))

usa_count = (df_2017['Q4: COUNTRY'] == 'U.S.A.').sum()
uk_count = (df_2017['Q4: COUNTRY'] == 'United Kingdom').sum()
canada_count = (df_2017['Q4: COUNTRY'] == 'Canada').sum()

print(f"\nTotal standardized as 'U.S.A.': {usa_count}")
print(f"Total standardized as 'United Kingdom': {uk_count}")
print(f"Total standardized as 'Canada': {canada_count}")

final_unique = df_2017['Q4: COUNTRY'].nunique()
print(
    f"Final unique country variations after US + UK + Canada cleanup: {final_unique}")

# 2. Create a hierarchical index using demographic fields from the 2016 survey
df_2016_indexed = df_2016.set_index([
    'Which country do you live in?',
    'Which state, province, county do you live in?',
    'Your gender:',
    'How old are you?'
])

print('\nHierarchical index created. Index levels:')
print(df_2016_indexed.index.names)

# 3. Extract survey question metadata from the 2017 PDF (manually
# reconstructed)
pdf_metadata = {
    'Question': ['Q1', 'Q2', 'Q3', 'Q4', 'Q5'],
    'Prompt': [
        'Are you actually going trick or treating yourself?',
        'Your gender',
        'How old are you?',
        'What country do you live in?',
        'Which state, province, county do you live in?'
    ]
}
df_pdf_meta = pd.DataFrame(pdf_metadata)
print('\nPDF metadata table created:')
print(df_pdf_meta)

# 4. Reshape the 2017 survey and merge with metadata and candy reference data

# Identify ID fields and candy question columns
id_vars = [
    'Internal ID',
    'Q1: GOING OUT?',
    'Q2: GENDER',
    'Q3: AGE',
    'Q4: COUNTRY',
    'Q5: STATE, PROVINCE, COUNTY, ETC'
]
value_vars = [col for col in df_2017.columns if col.startswith('Q6')]

# Melt the dataset from wide to long format
original_rows = len(df_2017)
df_2017_long = df_2017.melt(
    id_vars=id_vars,
    value_vars=value_vars,
    var_name='Question_Candy',
    value_name='Response'
)

melted_rows = len(df_2017_long)
print(f'\nOriginal 2017 rows: {original_rows}')
print(f'Rows after melting: {melted_rows}')
print(f'Rows created during melt: {melted_rows - original_rows}')

# Clean candy names for merging
df_2017_long['Candy'] = df_2017_long['Question_Candy'].str.replace(
    'Q6 \\| ', '', regex=True)

# Merge with candy metadata
df_merged = df_2017_long.merge(
    df_candydata,
    left_on='Candy',
    right_on='ITEM',
    how='left'
)

merged_rows = len(df_merged)
unmatched = df_merged['JOY'].isna().sum()

print(f'\nRows after merge: {merged_rows}')
print(f'Unmatched candy items (no metadata found): {unmatched}')
print('\nMerged dataset preview:')
print(df_merged.head())
```

### Chapter 10 — Grouping and Split/Apply/Combine

In this chapter, I focus on grouping operations using dictionaries, Series objects, custom functions, and index levels, as well as applying the split–apply–combine pattern to the reshaped candy survey dataset. These techniques make it possible to summarize sentiment, compare demographic patterns, and analyze response distributions across multiple dimensions.

### Steps:

1. Use dictionaries and Series objects to group candy items into broader categories (e.g., chocolate, fruit, novelty, non‑candy) and compute aggregated JOY, MEH, and DESPAIR values for each group.  
2. Apply custom functions to group respondents into age brackets and analyze how candy sentiment varies across age groups.  
3. Use index‑level grouping to summarize JOY, DESPAIR, and NET FEELIES using the hierarchical index created earlier in Chapter 8.  
4. Demonstrate the split–apply–combine pattern by computing grouped statistics across demographic fields such as gender, age, and country.  
5. Generate cross‑tabulations (crosstabs) to compare JOY/MEH/DESPAIR distributions across demographic categories and candy types.

<hr style="border: none; height: 3px; background-color: black;">

```python
# 1. Grouping with dictionaries and Series
# Map each candy item to a broader category (e.g., Chocolate, Fruit Candy)
candy_categories = {
    '100 Grand Bar': 'Chocolate',
    'Any full-sized candy bar': 'Chocolate',
    'Black Jacks': 'Licorice',
    'Bottle Caps': 'Sugar Candy',
    'Broken glow stick': 'Non-Candy',
    'Candy Corn': 'Sugar Candy',
    'Kit Kat': 'Chocolate',
    'Reese’s Peanut Butter Cups': 'Chocolate',
    'Skittles': 'Fruit Candy',
    'Starburst': 'Fruit Candy',
    'Swedish Fish': 'Fruit Candy'
}

category_series = pd.Series(candy_categories, name='Category')

# Add category labels to the merged dataset
df_merged['Category'] = df_merged['ITEM'].map(category_series)

# Compute average sentiment scores for each candy category
grouped_categories = df_merged.groupby(
    'Category')[['JOY', 'DESPAIR', 'NET FEELIES']].mean()

print('Grouped by candy category:')
print(grouped_categories.head())


# 2. Grouping with a custom function (age brackets)
# Convert raw ages into bracketed age groups
def age_bracket(age):
    try:
        age = int(age)
    except BaseException:
        return 'Unknown'
    if age < 18:
        return 'Under 18'
    elif age < 30:
        return '18–29'
    elif age < 50:
        return '30–49'
    else:
        return '50+'


df_merged['Age Bracket'] = df_merged['Q3: AGE'].apply(age_bracket)

# Compute average sentiment scores by age bracket
grouped_age = df_merged.groupby('Age Bracket')[
    ['JOY', 'DESPAIR', 'NET FEELIES']].mean()

print('\nGrouped by age bracket:')
print(grouped_age)


# 3. Grouping with index levels (gender + age)
# Use a multi-index to group by gender and age simultaneously
df_indexed = df_merged.set_index(['Q2: GENDER', 'Q3: AGE'])

grouped_index_levels = df_indexed.groupby(
    level=[0, 1])[['JOY', 'DESPAIR']].mean()

print('\nGrouped by gender and age (index levels):')
print(grouped_index_levels.head())


# 4. Split–apply–combine (mean JOY per candy per gender)
# Demonstrates the classic pattern: split data → apply function → combine
# results
split_apply = df_merged.groupby(['Q2: GENDER', 'ITEM'])['JOY'].mean()

print('\nSplit–apply–combine (JOY by gender and candy):')
print(split_apply.head())


# 5. Crosstab of JOY/MEH/DESPAIR by gender
# Shows the distribution of response types for each gender
crosstab_responses = pd.crosstab(
    df_merged['Q2: GENDER'],
    df_merged['Response'],
    normalize='index'
)

print('\nCrosstab of response distribution by gender:')
print(crosstab_responses)
```

### Chapter 11 — Working with Dates and Times

In this chapter, I demonstrate how to convert string fields into datetime objects, generate date ranges, work with frequencies and date offsets, and convert between timestamps and periods. These operations form the foundation of time‑series analysis and allow for more flexible indexing, grouping, and resampling.

### Steps for this section:

1. Convert year columns to numeric values to ensure they can be used in datetime operations.  
2. Create an annual `AccessionPeriod` to represent each year as a period rather than a timestamp.  
3. Apply date offsets such as `MonthEnd`, `YearBegin`, and `BusinessDay` to explore how timestamps shift under different offset rules.  
4. Convert timestamps to periods (quarter and decade) to demonstrate how pandas handles coarser time frequencies.  
5. Summarize the results and verify that the period conversions behave as expected.

<hr style="border: none; height: 3px; background-color: black;">

```python
# 1. Convert year columns to numeric
# Convert year-like strings into integers using a strict YYYY format
met['AccessionYear'] = pd.to_datetime(
    met['AccessionYear'], format='%Y', errors='coerce'
).dt.year.astype('Int64')

met['Object Begin Date'] = pd.to_datetime(
    met['Object Begin Date'], format='%Y', errors='coerce'
).dt.year.astype('Int64')

met['Object End Date'] = pd.to_datetime(
    met['Object End Date'], format='%Y', errors='coerce'
).dt.year.astype('Int64')

print('Converted year columns:')
print(met[['AccessionYear', 'Object Begin Date', 'Object End Date']].head())


# 2. Create an annual AccessionPeriod
# Reset index to avoid alignment issues when updating PeriodIndex
met = met.reset_index(drop=True)

# Extract valid years as strings for PeriodIndex construction
valid_years_str = met['AccessionYear'].dropna().astype(str)

# Create a PeriodIndex anchored to year-end (Dec 31)
period_index = pd.PeriodIndex(valid_years_str, freq='Y-DEC')

# Initialize a full-length period Series and update only valid rows
accession_period = pd.Series(pd.NaT, index=met.index, dtype='period[Y]')
accession_period.update(
    pd.Series(period_index, index=met['AccessionYear'].dropna().index)
)

met['AccessionPeriod'] = accession_period

# Convert periods back to timestamps for display
met['AccessionTimestamp'] = met['AccessionPeriod'].dt.to_timestamp(how='end')
met['AccessionTimestamp_Display'] = met['AccessionTimestamp'].dt.strftime(
    '%Y-%m-%d')

print('\nAccessionPeriod sample:')
print(met[['AccessionYear', 'AccessionPeriod',
      'AccessionTimestamp_Display']].head())


# 3. Apply date offsets
# Demonstrate MonthEnd, YearBegin, and BusinessDay offsets
example_date = pd.Timestamp('2020-01-15')

shift_month_end = example_date + pd.offsets.MonthEnd(1)
shift_year_begin = example_date + pd.offsets.YearBegin(1)
shift_business_day = example_date + pd.offsets.BusinessDay(5)

print('\nDate offsets:')
print('MonthEnd:', shift_month_end)
print('YearBegin:', shift_year_begin)
print('BusinessDay:', shift_business_day)


# 4. Convert timestamps to periods (quarter and decade)
# Convert Dec 31 timestamps to quarterly periods
met['AccessionQuarter'] = met['AccessionTimestamp'].dt.to_period('Q')

# Convert timestamps to decade periods using integer math
met['AccessionDecade'] = met['AccessionTimestamp'].apply(
    lambda ts: pd.Period(
        year=(ts.year // 10) * 10,
        freq='10Y'
    ) if pd.notna(ts) else pd.NaT
)

print('\nQuarter and Decade sample:')
print(met[['AccessionYear', 'AccessionPeriod',
           'AccessionQuarter', 'AccessionDecade']].head())


# 5. Summary
print('\nSummary:')
print('Total rows:', len(met))
print('Unique Accession Years:', met['AccessionYear'].dropna().nunique())
print('Earliest year:', met['AccessionYear'].min())
print('Latest year:', met['AccessionYear'].max())
print('Unique Quarters:', met['AccessionQuarter'].dropna().nunique())
print('Unique Decades:', met['AccessionDecade'].dropna().nunique())
```
